## Data Preparation
* Eliminate id and VIN - these are unique for the sample and won't show a relationship with price.
* Eliminate region - too many values and geographic information is somewhat included in the state column.
* Eliminate model - too many values to manage and probably too specific to develop a model. 
* Eliminate rows with null values. We're going to lose 75% of the data set if we eliminate size, but I think it is an important attribute. I'm going to keep size and reduce the data set size.

```python
# Step 1 - Eliminate columns
df = data.drop(['id', 'VIN', 'region', 'model'], axis=1)
df.info()
```

```python
# Remove Rows with Null Values
df = df.dropna()
df.shape
```

```python
# Look at relationship between price and year 
```
```python
price2['odometer^2'] = price2['odometer']**2
sns.scatterplot(data=price2, x = 'odometer', y='price')
plt.yscale('log')
plt.ylabel('Price $')
plt.xlabel('Odometer Reading')
plt.title('Relationship of Odometer Reading and Vehicle Price')
plt.savefig('images/odometer');
```

Above I explored the relationships between price, odometer and year. I tried logarithmic transformations of all the variables and the square of the odometer and year.
```python
# Delete rows with price = 0 and price > 1000000
df = df[df['price'] > 0]
df = df[df['price'] < 100000]
df.info
```

```python
# Get dummies of all the object features
df = pd.get_dummies(df)
df.info()
```

## Modeling

#### Step 1: Pull out a sample to test the model - not for training or cross validation

```python
# Select 10% of data for test set 
# Shuffle the data
set_seed = 42
df_shuffle = shuffle(df)

# Split the data
model_data, test_data = np.split(df_shuffle, [69023])
test_data.shape
```

```python
# Split the model data into X and y
Xtrain = model_data.drop(['price'],axis =1)
ytrain = model_data['price']
```

#### Step 2: Feature Selection - LASSO and Ridge to identify most important features. Use Cross Validation and GridSearch.

```python
# Assess independence of variables
corr = Xtrain.corr()
sns.heatmap(corr)
plt.title('Heatmap showing Correlation of Predictors')
plt.savefig('images/heatmap', bbox_inches='tight');
```

##### Perform Ridge Regression to Eliminate Features

```python
# Scale the model data
ss = StandardScaler()
Xtrain_scaled = pd.DataFrame(ss.fit_transform(Xtrain), columns=ss.get_feature_names_out())
```

```python
# Create a dictionary of alpha values 
params_dict = {'alpha': [1,10,100,500,600,625,650,660,670,680,700,800,900,1000]}

# Define ridge regression as the estimator
ridge = Ridge()

# Create a grid search using 5-fold Cross Validation and neg MSE to compare
grid = GridSearchCV(ridge, param_grid=params_dict, cv=5, scoring = 'neg_mean_squared_error')
grid.fit(Xtrain_scaled,np.log(ytrain))
best_alpha = grid.best_params_

print(f'Best alpha: {list(best_alpha.values())[0]}')
```

```python
# Find Coefficients for best fit model
best_model = grid.best_estimator_
ridge_coef = best_model.coef_
Features = list(Xtrain_scaled.columns.values)
ridge_coef_df = pd.DataFrame({'Features':Features, 'Ridge Coefficients':ridge_coef})

sns.barplot(data=ridge_coef_df, x='Features', y='Ridge Coefficients')
plt.tick_params(labelbottom=False)
plt.title('Ridge Coefficients Using Optimum Alpha =670')
plt.savefig('images/ridge_coef');
```

```python
# Eliminate features with coefficients near zero

ridge_coef_df.describe()
ridge_coef_reduced = ridge_coef_df[(ridge_coef_df['Ridge Coefficients'] >= 0.05) | (ridge_coef_df['Ridge Coefficients'] <= -0.05)]
ridge_coef_reduced['Features'].unique()
```

```python
# Reduce data set to 23 featurs found from Ridge Regression
feature_list = list(ridge_coef_reduced['Features'].unique())

Xtrain_reduced = Xtrain[[i for i in Xtrain.columns if i in feature_list]]
Xtrain_reduced.info()
```

##### Use sequential feature selection on the 23 features selected above to determine the optimum number of features

```python
# Use Linear Regression, apply a Grid Search to determine the optimum number of features.
# Note, I search n = 1-23, but completed in intervals only n=16-23 are shown below.
selector = SequentialFeatureSelector(estimator = LinearRegression())
pipe = Pipeline([
    ('SFS', selector),
    ('lin_reg', LinearRegression())
])

features = [16,17,18,19,20,21,22,23]
grid2 = GridSearchCV(pipe, {'SFS__n_features_to_select':features},
                   cv=5,
                   scoring='neg_mean_squared_error')

grid2.fit(Xtrain_reduced,np.log(ytrain))
best_n = grid2.best_params_

print(f'Best n: {list(best_n.values())[0]}')
```

```python
# Use the best run to select features
best_lr = grid2.best_estimator_
lr_list = list(best_lr.named_steps['SFS'].get_feature_names_out())
lr_list
```

```python
# Create a training data set with these 21 variables
Xtrain_lr = Xtrain_reduced.drop(['condition_excellent', 'transmission_manual'], axis=1)
Xtrain_lr.info()
```

##### Apply Lasso on the data set reduced to 23 columns.

```python
# Create a dictionary of alpha values 
params_dict = {'lasso__alpha': [0.1,1,10,100,1000]}

# Define lasso regression as the estimator
pipe = Pipeline([('scaler', StandardScaler()),
                     ('lasso', Lasso(random_state = 42))])

# Create a grid search using 5-fold Cross Validation and neg MSE to optimize alpha
grid3 = GridSearchCV(pipe, param_grid=params_dict, cv=5, scoring = 'neg_mean_squared_error')
grid3.fit(Xtrain_reduced,np.log(ytrain))
best_alpha2 = grid3.best_params_

print(f'Best alpha: {list(best_alpha2.values())[0]}')

```

```python
# Find coefficients from the best LASSO model

best_lasso = grid3.best_estimator_
lasso_coef = best_lasso.named_steps['lasso'].coef_
Features = list(Xtrain_reduced.columns.values)
lasso_coef_df = pd.DataFrame({'Features':Features, 'Lasso Coefficients':lasso_coef})
lasso_coef_df
```

##### Eliminate variables that went to zero

```python
# Remove all columns where coefficients = zero
lasso_list = list(lasso_coef_df.loc[lasso_coef_df["Lasso Coefficients"]==0.0,'Features'])
Xtrain_lasso = Xtrain_reduced.drop(lasso_list, axis=1)
Xtrain_lasso.info()
```

#### Step 3: Build a LInear Regression Model 

```python
# Train the model using the full dataset (minus the 10% held to test the model) = 'Model 1'
linreg = LinearRegression()
linreg.fit(Xtrain_lr,np.log(ytrain))
print(linreg.intercept_, linreg.coef_, linreg.score(Xtrain_lr,np.log(ytrain)))
```

```python
# Calculate adjusted R2 for model 1
M1adjr2 = 1-(1-linreg.score(Xtrain_lr,np.log(ytrain)))*(len(Xtrain_lr)-1)/(len(Xtrain_lr)-21-1)
M1adjr2
```

```python
# Train the model using Ridge and the full dataset
ridge = Ridge(alpha=760)
ridge.fit(Xtrain_lr,np.log(ytrain))
print(ridge.intercept_, ridge.coef_, ridge.score(Xtrain_lr,np.log(ytrain)))
```

```python
# Train the model using LR and the lasso dataset - 'Model 2'
linreg2 = LinearRegression()
linreg2.fit(Xtrain_lasso,np.log(ytrain))
print(linreg2.intercept_, linreg2.coef_, linreg2.score(Xtrain_lasso,np.log(ytrain)))
```

```python
# Calculate adjusted R2 for model 2
M2adjr2 = 1-(1-linreg2.score(Xtrain_lasso,np.log(ytrain)))*(len(Xtrain_lasso)-1)/(len(Xtrain_lasso)-21-1)
M2adjr2
```

## Evaluation

```python
# Evaluate the model using the test set
# Reduce test set to 21 features
test_final_y = test_data['price']

# Final test with Xtrain_lr columns
final_list = list(Xtrain_lr.columns.values)
test_final_X = test_data[[i for i in test_data.columns if i in final_list]]
test_final_X.info()

# Final test with Xtrain_lasso columns
final_list2 = list(Xtrain_lasso.columns.values)
test_final_X2 = test_data[[i for i in test_data.columns if i in final_list2]]
```

```python
# Calculate MSE for Model 1
linreg = LinearRegression()
linreg.fit(Xtrain_lr,np.log(ytrain))
testpreds = linreg.predict(test_final_X)
trainpreds = linreg.predict(Xtrain_lr)
testmse1 = mean_squared_error(testpreds, np.log(test_final_y))
trainmse1 = mean_squared_error(trainpreds, np.log(ytrain))
print(testmse1, trainmse1)
```

```python
# Calculate MSE for Model 1
linreg2 = LinearRegression()
linreg2.fit(Xtrain_lasso,np.log(ytrain))
testpreds2 = linreg2.predict(test_final_X2)
trainpreds2 = linreg2.predict(Xtrain_lasso)
testmse2 = mean_squared_error(testpreds2, np.log(test_final_y))
trainmse2 = mean_squared_error(trainpreds2, np.log(ytrain))
print(testmse2, trainmse2)
```

```python
# Create a table to show results
results = [['Model 1', linreg.score(Xtrain_lr,np.log(ytrain)),M1adjr2,trainmse1,testmse1],
           ['Model 2',linreg2.score(Xtrain_lasso,np.log(ytrain)),M2adjr2,trainmse2,testmse2]]
table = pd.DataFrame(results, columns=['Model', 'R-squared', 'Adj R-squared', 'Training MSE', 'Test MSE'])
table.style.hide_index()
```

```python
# Create a table showing the coefficients of Model 1
M1_coef = linreg.coef_
M1_features = list(Xtrain_lr.columns.values)
M1_table = pd.DataFrame({'Features': M1_features, 'Model Coefficients': M1_coef})
M1_table.style.hide_index()
```
